**Ahora, comenzaremos a trabajar con los resultados electorales de las elecciones generales de 2023**

```python
resultados = pd.read_csv('/content/drive/MyDrive/AA DATOS/ResultadosElectorales_2023.csv', low_memory=False)
```

**Limpiamos la columna secion_nombres para que quede igual a la de nuestro df de circuitos electorales ya que a partir de aqui vamos a poder fusionar las dos df**

```python
resultados['seccion_nombre'] = circuitos['seccion_nombre'].str.replace('Comuna ', '', "0").str.lstrip('0')
```

```python
resultados
```

```python
resultados1 = pd.merge(circuitos, resultados, on='seccion_nombre', how='inner')
```

```python
resultados1
```

**Para nuestro analisis decidimos filtrar primero, como fueron las elecciones de Juntos por el Cambio, el partido que mayor cantidad de votos sacó**

```python
resultados2 = resultados1[(resultados1['cargo_nombre'] == 'PRESIDENTE Y VICE') & (resultados1['agrupacion_nombre'] == 'JUNTOS POR EL CAMBIO')]

resultados2
```

```python
import geopandas as gpd
from shapely import wkt
import matplotlib.pyplot as plt

# Supongamos que resultados_agrupados es tu DataFrame agrupado

# Convertir la columna 'WKT' a geometrías
resultados2['geometry'] = resultados2['WKT'].apply(wkt.loads)

# Crear un GeoDataFrame
gdf = gpd.GeoDataFrame(resultados2, geometry='geometry')

# Definir un rango para la escala de colores (ajusta según tus datos)
vmin = resultados2['votos_cantidad'].min()
vmax = resultados2['votos_cantidad'].max()

# Plotear el mapa de calor con paleta de colores azules más fuertes y suaves
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(column='votos_cantidad', cmap='Wistia', legend=True, legend_kwds={'label': "Votos Cantidad"}, ax=ax, vmin=vmin, vmax=vmax)
plt.title('Mapa de cantidad de votos de Juntos por el Cambio')

# Eliminar etiquetas de los ejes x e y
ax.set_xticks([])
ax.set_yticks([])

plt.show()
```

**Será importante conocer como fueron los votos a JxC en cada comuna ya que nos permitirá ver donde debemos ir a buscar los votos para competir contra UxP en el ballotage**

**Una alianza con JxC y movilizar la campaña hacia las comunas donde obtuvieron una mayor cantidad de votos, sería el factor fundamental para continuar con la campaña**

**Un factor importante con el cual debemos trabajar es con la cuestión del genero. En CABA, hay una notoria paridad y nuestro partido no cuenta con el apoyo femenino. Esto queda demostrado al comparar nuestros resultados con el siguiente grafico**

```python

import geopandas as gpd
import matplotlib.pyplot as plt
from shapely import wkt

# Ruta al archivo CSV
ruta_archivo = '/content/drive/MyDrive/AA DATOS/circuitos-electorales.csv'

# Intentar abrir el archivo CSV con Pandas y convertirlo a GeoDataFrame

    # Lee el archivo CSV con Pandas
circuitos = pd.read_csv(ruta_archivo)

    # Convierte el DataFrame de Pandas a un GeoDataFrame
gdf = gpd.GeoDataFrame(circuitos, geometry=gpd.GeoSeries(circuitos['WKT'].apply(wkt.loads)))

    # Calcula la diferencia entre la población masculina y femenina
gdf['Diferencia_HM'] = gdf['MASCULINO'] - gdf['FEMENINO']

    # Configura la visualización
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

    # Muestra la diferencia en un mapa divergente
gdf.plot(column='Diferencia_HM', ax=ax, legend=True, cmap='coolwarm', legend_kwds={'label': "Diferencia de Población (Hombres - Mujeres)"})

    # Añade etiquetas y título
ax.set_title("Diferencia de Población según Hombres - Mujeres por Barrio en Buenos Aires")
ax.set_xlabel("Longitud")
ax.set_ylabel("Latitud")

plt.show()
```
