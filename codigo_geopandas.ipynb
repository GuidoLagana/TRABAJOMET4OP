```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
circuitos = gpd.read_file('/content/drive/MyDrive/AA DATOS/circuitos-electorales.csv')
circuitos
```

```python
circuitos= circuitos.rename(columns={'COMUNA': 'seccion_nombre'})
```

```python
circuitos
```

```python
pd.set_option('mode.chained_assignment', None)
```

```python
resultados = pd.read_csv('/content/drive/MyDrive/AA DATOS/ResultadosElectorales_2023.csv', low_memory=False)
```

```python
resultados['seccion_nombre'] = circuitos['seccion_nombre'].str.replace('Comuna ', '').str.lstrip('0')
```

```python
resultados
```

```python
resultados1 = pd.merge(circuitos, resultados, on='seccion_nombre', how='inner')
```

```python
resultados1
```

```python
resultados2 = resultados1[(resultados1['cargo_nombre'] == 'PRESIDENTE Y VICE') & (resultados1['agrupacion_nombre'] == 'LA LIBERTAD AVANZA')]

resultados2
```

```python
agrupados = resultados2.groupby(['WKT', 'ID', 'CIRCUITO_N', 'seccion_nombre', 'BARRIO', 'PERIMETRO', 'AREA', 'agrupacion_nombre']).agg({'votos_cantidad': 'sum'}).reset_index()
```

```python
import geopandas as gpd
from shapely import wkt
import matplotlib.pyplot as plt

# Supongamos que agrupados es tu DataFrame agrupado

# Convertir la columna 'WKT' a geometrías
agrupados['geometry'] = agrupados['WKT'].apply(wkt.loads)

# Crear un GeoDataFrame
gdf = gpd.GeoDataFrame(agrupados, geometry='geometry')

# Plotear el mapa de calor
gdf.plot(column='votos_cantidad', cmap='viridis', legend=True, legend_kwds={'label': "Votos Cantidad"}, figsize=(10, 10))
plt.title('Mapa de Calor de Votos Cantidad por Área Geográfica')
plt.show()
```

```python
import geopandas as gpd
from shapely import wkt
import matplotlib.pyplot as plt

# Supongamos que resultados_agrupados es tu DataFrame agrupado

# Convertir la columna 'WKT' a geometrías
agrupados['geometry'] = agrupados['WKT'].apply(wkt.loads)

# Crear un GeoDataFrame
gdf = gpd.GeoDataFrame(agrupados, geometry='geometry')

# Definir un rango para la escala de colores (ajusta según tus datos)
vmin = agrupados['votos_cantidad'].min()
vmax = agrupados['votos_cantidad'].max()

# Plotear el mapa de calor con paleta de colores azules más fuertes y suaves
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(column='votos_cantidad', cmap='Purples', legend=True, legend_kwds={'label': "Votos Cantidad"}, ax=ax, vmin=vmin, vmax=vmax)
plt.title('Mapa de Calor de Votos Cantidad por Área Geográfica (Azules Fuertes y Suaves)')

# Eliminar etiquetas de los ejes x e y
ax.set_xticks([])
ax.set_yticks([])

plt.show()
```
